```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("world population year wise.csv")
```

```python
population_2021 = df["2021"].dropna()
```

```python
plt.hist(population_2021, bins=30)
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.title("Distribution of Country Populations (2021)")
plt.show()
```

```python
bins = [0, 10_000_000, 50_000_000, 100_000_000, 1_500_000_000]
labels = ["<10M", "10–50M", "50–100M", ">100M"]

df["Population Category"] = pd.cut(df["2021"], bins=bins, labels=labels)
```

```python
category_counts = df["Population Category"].value_counts()
```

```python
category_counts.plot(kind="bar")
plt.xlabel("Population Category")
plt.ylabel("Number of Countries")
plt.title("Country Count by Population Category (2021)")
plt.show()
```

```python
genders = ["Male", "Female", "Male", "Female", "Male", "Female", "Female"]
```

```python
from collections import Counter
gender_count = Counter(genders)
```

```python
plt.bar(gender_count.keys(), gender_count.values())
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution in Population")
plt.show()
```
